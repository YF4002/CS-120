For Project 3, create an application that accepts the data for any number of a student’s courses, which includes each course’s:
* title
* credit hours
* numerical score

The application will 
* assign a letter grade to each course based on the score 
* assign quality points based on the letter grade
* calculate the GPA
    1. assign quality points based on letter grades:
        * A is 4 points
        * B is 3 points
        * C is 2 points
        * D is 1 point
        * F is 0 points
    2. For each course, multiply quality points by the credit hours, then sum all of these results to get total points
        * Example: total_points = (course_1_points * course_1_hours) + (course_2_points * course_2_hours) + (course_3_points * course_3_hours) + ... etc. 
    3. Sum the credit hours of all the classes
    4. Divide these total points by the total credit hours to get the gpa         
* display each course’s data (including grade and points)
* display the student’s GPA

## Project 3

Project 3 has **6 functions**.<br>
Use Project 2 for the first 4 functions.

### These 2 functions should stay the same as in Project 2:
**1. enter_courses()**
* start with an empty list for courses
* prompt student to enter course title or q to quit
* if student does not quit
    * prompt student for course credit hours
    * prompt student for course numerical score
    * use each of these prompts (course, hours, score) for keys to a dictionary for that course (each course is a separate dictionary)
    * use the student’s inputs as the values to the keys
    * add the course dictionary to the course list
* if the student quits
* return the course list

**2. assign_letter_grade()**
* takes a course list as a parameter
* for each course dictionary in the list
    * compare the course score with percentages to determine the letter grade
        * example: 90 – 100 is an A, 80 – 89 is a B, etc.
    * use grade as a key and the assigned letter grade as its value to add another key/value pair to the course dictionary
* after assigning a letter grade to each course in the list, return the course list

### Modify these functions by adding the items bolded below:
**3. display_course_info()**
* takes a course list and **gpa** as a parameters
* prints a Courses header
* prints all the key/value pairs for each course on a separate line
* **prints the GPA**
* see Output Example

**4. main()**
* calls enter_courses() and assigns the returned course list to a variable
* calls assign_letter_grades(), with the course list as its argument, and assigns the returned list to a variable
* **calls assign_points(), with the course list as its argument, and assigns the returned list to a variable**
* **calc_gpa(), with the course list as its argument, and assigns the returned *value* to a variable (gpa)**
* calls display_course_info(), with the course list **and gpa** as its argument 

### The following 2 functions need to be added to the project:
**5. assign_points()**
* takes a course list as a parameter
* for each course dictionary in the list
    * compare the course letter_grade value with letters (A to F) to determine the quality points (if the letter_grade is "A", then assign 4 to a variable (points), etc.)
        * "A" is 4 points
        * "B" is 3 points
        * "C" is 2 points
        * "D" is 1 point
        * "F" is 0 points
    * use "points" as a key and the assigned quality points as its value to add another key/value pair to the course dictionary
* after assigning a point key/value pair to each course in the list, return the course list

6. **calculate_gpa()**
* takes a course list as a parameter
* use a variable to keep track of total quality points
* use a variable to keep track of total credit hours
* for each course dictionary in the list
    * multiply the course's points by the credit hours, add the result to total points variable
    * add the course's hours to the total hours variable
* calculate the gpa by dividing total quality points by the total hours, assign this to a gpa variable
* return gpa
